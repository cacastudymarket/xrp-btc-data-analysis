```python
import pandas as pd

btc = pd.read_csv("Data/Bitfinex_BTCUSD_d.csv", skiprows=1)
xrp = pd.read_csv("Data/Bitfinex_XRPUSD_d.csv", skiprows=1)

btc.head()
```

```python
# Kolom 'date' jadi datetime dan urutkan dari tanggal lama ke baru
btc['date'] = pd.to_datetime(btc['date'])
btc = btc.sort_values('date')

xrp['date'] = pd.to_datetime(xrp['date'])
xrp = xrp.sort_values('date')

# Cek data terakhir
print("BTC Terbaru:", btc.iloc[-1]['date'], "=", btc.iloc[-1]['close'])
print("XRP Terbaru:", xrp.iloc[-1]['date'], "=", xrp.iloc[-1]['close'])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
plt.plot(btc['date'], btc['close'], label='BTC/USD', color='orange')
plt.plot(xrp['date'], xrp['close'], label='XRP/USD', color='blue')
plt.title('Crypto Price Comparison')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import datetime

# Tanggal peristiwa penting
btc_event = datetime.datetime(2021, 9, 7)  # BTC Legal Tender in El Salvador

# Grafik BTC dengan garis event
plt.figure(figsize=(12, 5))
plt.plot(btc['date'], btc['close'], label='BTC/USD', color='orange')
plt.axvline(btc_event, color='red', linestyle='--', label='BTC Legalized in El Salvador (2021)')
plt.title('BTC Price with Event Marker')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Tanggal estimasi adopsi nasional XRP di Jepang (perkiraan 2025)
xrp_event = datetime.datetime(2025, 1, 1)

# Grafik XRP dengan garis event
plt.figure(figsize=(12, 5))
plt.plot(xrp['date'], xrp['close'], label='XRP/USD', color='blue')
plt.axvline(xrp_event, color='green', linestyle='--', label='Planned Nationwide XRP Adoption in Japan (2025)')
plt.title('XRP Price with Event Marker')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.legend()
plt.grid(True)
plt.show()

```
